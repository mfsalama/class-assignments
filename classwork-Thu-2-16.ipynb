```python
def makeBirdDict(filename):
    #this function opens the data file from Thomas et al and extracts the species and families as dictionary 
    birdDict = {}
    #open the file
    fobj = open("/home/eeb177-student/Desktop/eeb-177/Homework/formatted.eBird2.csv", "r", encoding = "ISO-8859-15")
    #read the file (and store list)
    all_records = fobj.readlines()[1:]
    for line in all_records:
        #record_elements = line.split(",")
        species = line.split(",")[0]
        family= line.split(",")[2]
        birdDict[species] = family
    #split lines on comma
    #stick species and family into birdDict
    
    return birdDict
```

```python
thomas_birds = makeBirdDict("/home/eeb177-student/Desktop/eeb-177/Homework/formatted.eBird2.csv")
species = list (thomas_birds.keys())
species.sort()
first_species = species[0]
first_family = thomas_birds[first_species]
print( first_species, first_family)
print("family of" + species[0] + "is" + thomas_birds[species[0]])
```
